# 도로 사고 통계량(인천) 과 변수 분석

```python
import requests
import os
import dotenv
import pandas as pd
import time

dotenv.load_dotenv()
authKey = os.getenv('PUBLIC_DATA_API_KEY')
years = ['2019', '2020', '2021', '2022', '2023', '2024','2025']
sido = '28'   # 인천광역시 코드

# 1. 인천광역시의 시군구 코드 리스트 (중구, 동구, 미추홀구, 연수구, 남동구, 부평구, 계양구, 서구, 강화군, 옹진군)
gu_codes = ['110', '140', '177', '185', '200', '237', '245', '260', '710', '720']
all_data = [] # 각 구별 데이터를 담을 리스트

for year in years:
    print(f"📅 {year}년 데이터 수집 중...")
    
    for gu in gu_codes:
        url = f'https://opendata.koroad.or.kr/data/rest/frequentzone/lg?authKey={authKey}&searchYearCd={year}&sido={sido}&guGun={gu}&type=json&numOfRows=100'
        
        try:
            response = requests.get(url)
            data = response.json()
            
            if data.get('items') and data['items'].get('item'):
                items = data['items']['item']
                for item in items:
                    item['year'] = year
                
                all_data.extend(items)
                print(f"   ✅ {year}년 구 코드 {gu}: {len(items)}건 완료")
            else:
                print(f"   ⚪ {year}년 구 코드 {gu}: 데이터 없음")
        except Exception as e:
            print(f"   ❌ {year}년 구 코드 {gu} 에러: {e}")
            

if all_data:
    incheon_all_df = pd.DataFrame(all_data)
    print("-" * 50)
    print(f"✨ 총 {len(incheon_all_df)}건 수집 완료!")
else:
    print("수집된 데이터가 없습니다.")
```

```python
## 화물차

import requests
import os
import dotenv
import pandas as pd
import time

dotenv.load_dotenv()
authKey = os.getenv('PUBLIC_DATA_API_KEY')
years = ['2020','2021', '2022', '2023', '2024','2025']
sido = '28'

# 1. 인천광역시의 시군구 코드 리스트 (중구, 동구, 미추홀구, 연수구, 남동구, 부평구, 계양구, 서구, 강화군, 옹진군)
gu_codes = ['110', '140', '177', '185', '200', '237', '245', '260', '710', '720']
all_data = [] # 각 구별 데이터를 담을 리스트

for year in years:
    print(f"📅 {year}년 데이터 수집 중...")
    
    for gu in gu_codes:
        url = f'https://opendata.koroad.or.kr/data/rest/frequentzone/truck?authKey={authKey}&searchYearCd={year}&sido={sido}&guGun={gu}&type=json&numOfRows=100'
        
        try:
            response = requests.get(url)
            data = response.json()
            
            if data.get('items') and data['items'].get('item'):
                items = data['items']['item']
                # 각 아이템에 '수집연도' 정보를 추가해주면 나중에 분석하기 좋습니다.
                for item in items:
                    item['year'] = year 
                
                all_data.extend(items)
                print(f"   ✅ {year}년 구 코드 {gu}: {len(items)}건 완료")
            else:
                print(f"   ⚪ {year}년 구 코드 {gu}: 데이터 없음")
        except Exception as e:
            print(f"   ❌ {year}년 구 코드 {gu} 에러: {e}")
            

if all_data:
    incheon_truck = pd.DataFrame(all_data)
    print("-" * 50)
    print(f"✨ 총 {len(incheon_truck)}건 수집 완료! (2019~2024)")
else:
    print("수집된 데이터가 없습니다.")
```

```python
create_table_query1 = """
            CREATE TABLE IF NOT EXISTS incheon_accidents (
                afos_id VARCHAR(50),
                afos_fid VARCHAR(50) PRIMARY KEY,
                sido_sgg_nm VARCHAR(100),
                spot_nm VARCHAR(255),
                occrrnc_cnt INT,
                caslt_cnt INT,
                dth_dnv_cnt INT,
                se_dnv_cnt INT,
                sl_dnv_cnt INT,
                lo_crd DOUBLE,
                la_crd DOUBLE
            ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4;
            """

create_table_query2 = """
            CREATE TABLE IF NOT EXISTS incheon_truck (
                afos_id VARCHAR(50),
                afos_fid VARCHAR(50) PRIMARY KEY,
                sido_sgg_nm VARCHAR(100),
                spot_nm VARCHAR(255),
                occrrnc_cnt INT,
                caslt_cnt INT,
                dth_dnv_cnt INT,
                se_dnv_cnt INT,
                sl_dnv_cnt INT,
                lo_crd DOUBLE,
                la_crd DOUBLE
            ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4;
            """

insert_query1 = """
            INSERT IGNORE INTO incheon_accidents 
            (afos_id, afos_fid, sido_sgg_nm, spot_nm, occrrnc_cnt, caslt_cnt, dth_dnv_cnt, se_dnv_cnt, sl_dnv_cnt, lo_crd, la_crd)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
            """

insert_query2 = """
            INSERT IGNORE INTO incheon_truck 
            (afos_id, afos_fid, sido_sgg_nm, spot_nm, occrrnc_cnt, caslt_cnt, dth_dnv_cnt, se_dnv_cnt, sl_dnv_cnt, lo_crd, la_crd)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
            """
```

```python
## 테이블 생성 및 적재 (사고량)

import mysql.connector as mc
import os
import pandas as pd
from mysql.connector import Error
from datetime import datetime, timedelta
dotenv.load_dotenv(override=True)

DB_CONFIG = {
        'host': os.getenv('DB_HOST'),
        'port': int(os.getenv('DB_PORT', 3306)),
        'user': os.getenv('DB_USER'),
        'password': os.getenv('DB_PASSWORD'),
        'database': os.getenv('DB_DATABASE'),
        'autocommit': False,
}

def save_to_mysql(df,t_query , i_query):
    try:
        conn = mc.connect(**DB_CONFIG)
        cursor = conn.cursor()
        if conn.is_connected():
            cursor = conn.cursor()

            cursor.execute(t_query)
            conn.commit()
            print("✅ 테이블 생성(또는 확인) 완료")
            # 3. 데이터 적재 (Bulk Insert)
            
            cols = ['afos_id', 'afos_fid', 'sido_sgg_nm', 'spot_nm', 
                    'occrrnc_cnt', 'caslt_cnt', 'dth_dnv_cnt', 'se_dnv_cnt', 
                    'sl_dnv_cnt', 'lo_crd', 'la_crd']
            
            # 전처리: 숫자로 변환이 필요한 컬럼들 처리
            for col in ['occrrnc_cnt', 'caslt_cnt', 'dth_dnv_cnt', 'se_dnv_cnt', 'sl_dnv_cnt']:
                df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
            df['lo_crd'] = pd.to_numeric(df['lo_crd'], errors='coerce')
            df['la_crd'] = pd.to_numeric(df['la_crd'], errors='coerce')
            # 데이터를 튜플 형태로 변환
            data_to_insert = [tuple(x) for x in df[cols].values]

            # executemany로 한 번에 밀어넣기 (성능 최적화)
            cursor.executemany(i_query, data_to_insert)
            conn.commit()
            print(f"✨ {cursor.rowcount}개의 행이 성공적으로 적재되었습니다.")
    except Error as e:
        print(f"❌ 에러 발생: {e}")
    finally:
        if conn.is_connected():
            cursor.close()
            conn.close()
            print("🔌 DB 연결 종료")

save_to_mysql(incheon_all_df,create_table_query1,insert_query1)
save_to_mysql(incheon_truck,create_table_query2,insert_query2)
```

## Pyplot graph

```python
import plotly.express as px

query = """
SELECT
    a.sido_sgg_nm AS region, -- 지역명
    LEFT(a.afos_id, 4) AS year, -- 연도
    SUM(a.occrrnc_cnt) AS total_count, -- 전체 사고건수
    SUM(t.occrrnc_cnt) AS truck_count, -- 화물차 사고건수
    ROUND((SUM(t.occrrnc_cnt) / SUM(a.occrrnc_cnt)) * 100, 2) AS truck_ratio, -- 화물차 비중
    SUM(t.dth_dnv_cnt) AS death_count -- 화물차 사망자수
FROM incheon_accidents a
JOIN incheon_truck t
    ON a.sido_sgg_nm = t.sido_sgg_nm 
    AND LEFT(a.afos_id, 4) = LEFT(t.afos_id, 4)
GROUP BY a.sido_sgg_nm, year
ORDER BY year DESC, truck_ratio DESC;
"""

try:
    conn = mc.connect(**DB_CONFIG)
    df = pd.read_sql(query, conn)
    print("✅ 분석 데이터 로드 완료!")
finally:
    if conn.is_connected():
        conn.close()

fig = px.line(df,
              x="year", 
              y="truck_ratio", 
              color="region", # 'region' 컬럼 사용
              hover_data=["total_count", "truck_count", "death_count"], 
              markers=True,
              title="인천광역시 구별 연도별 화물차 사고 비중 추이")

fig.update_layout(
    xaxis_title="연도",
    yaxis_title="화물차 사고 비중 (%)",
    legend_title="지역구",
    template="plotly_white"
)

fig.show()
```
